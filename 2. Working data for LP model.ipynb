```python
import pandas as pd
import numpy as np
```

This notebook will be used to create working data sets and .csv files for use in the LP optimization model.

```python
# Read .csv
log = pd.read_csv('data/Sim data/sim_log_unlcap.csv', index_col=[0])
```

```python
log
```

```python
df_lp = pd.DataFrame()
```

```python
df_lp[['Time', 'Simulation', 'Depalletizer']] = log[['Time', 'Simulation', 'Actual speed']][(log['Machine'] == 'Depalletizer')].reset_index(drop = True)
df_lp['Filler'] = log['Actual speed'][(log['Machine'] == 'Filler')].reset_index(drop = True)
df_lp['Pasteurizer'] = log['Actual speed'][(log['Machine'] == 'Pasteurizer')].reset_index(drop = True)
df_lp['Labeler'] = log['Actual speed'][(log['Machine'] == 'Labeler_1')].reset_index(drop = True) + log['Actual speed'][(log['Machine'] == 'Labeler_2')].reset_index(drop = True)
df_lp['GPI'] = log['Actual speed'][(log['Machine'] == 'GPI')].reset_index(drop = True)
df_lp['SN64'] = log['Actual speed'][(log['Machine'] == 'SN64')].reset_index(drop = True)
df_lp['Palletizer'] = log['Actual speed'][(log['Machine'] == 'Palletizer_1')].reset_index(drop = True) + log['Actual speed'][(log['Machine'] == 'Palletizer_2')].reset_index(drop = True)
```

```python
df_lp
```

```python
df_lp.to_csv('data/LP data/df_lp.csv', index = False)
```

```python
df_sim1 = df_lp[df_lp['Simulation'] == 1]
df_sim1.to_csv('data/LP data/sim1.csv', index = False)
```
