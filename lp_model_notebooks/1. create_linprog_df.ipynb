```python
import pandas as pd
import numpy as np
```

This notebook will be used to create working data sets and .csv files for use in the LP optimization model.

```python
# Read .csv
simlog_df = pd.read_csv('../data/sim_log.csv', index_col=[0])
simlog_df.head()
```

```python
# Create an empty DataFrame
linprog_df = pd.DataFrame()
# Add the correct columns for `Time`, `Simulation` and all of the machines
# The rows are filtered for each machine.
# The index of the resulting DataFrame is reset.
linprog_df[['Time', 'Simulation', 'Depalletizer']] = simlog_df[['Time', 'Simulation', 'Actual speed']][(simlog_df['Machine'] == 'Depalletizer')].reset_index(drop = True)
linprog_df['Filler'] = simlog_df['Actual speed'][(simlog_df['Machine'] == 'Filler')].reset_index(drop = True)
linprog_df['Pasteurizer'] = simlog_df['Actual speed'][(simlog_df['Machine'] == 'Pasteurizer')].reset_index(drop = True)
linprog_df['Labeler'] = simlog_df['Actual speed'][(simlog_df['Machine'] == 'Labeler_1')].reset_index(drop = True) + simlog_df['Actual speed'][(simlog_df['Machine'] == 'Labeler_2')].reset_index(drop = True)
linprog_df['Capper'] = simlog_df['Actual speed'][(simlog_df['Machine'] == 'Capper')].reset_index(drop = True)
linprog_df['Packer'] = simlog_df['Actual speed'][(simlog_df['Machine'] == 'Packer')].reset_index(drop = True)
linprog_df['Palletizer'] = simlog_df['Actual speed'][(simlog_df['Machine'] == 'Palletizer_1')].reset_index(drop = True) + simlog_df['Actual speed'][(simlog_df['Machine'] == 'Palletizer_2')].reset_index(drop = True)
# Print first 5 rows
linprog_df.head()
```

```python
# Save to .csv
linprog_df.to_csv('../data/linprog_df.csv', index = False)
```
